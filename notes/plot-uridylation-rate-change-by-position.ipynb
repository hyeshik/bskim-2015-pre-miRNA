```python
ctlreads = pd.read_table('../stats/Control.mods.txt.gz', compression='gzip').fillna('')
kdreads = pd.read_table('../stats/TUT247KD.mods.txt.gz', compression='gzip').fillna('')
```

```python
def is_utail(seq):
    return set(seq.upper()) == set('T')
is_utail(''), is_utail('T'), is_utail('TTT'), is_utail('ATT')
```

```python
def make_modrate_tbl_by_pos(reads):
    pos_counts = []

    for hairpin, grp in reads.groupby('hairpin'):
        for endpos, subgrp in grp.groupby('refEndPos'):
            utailed = subgrp['addedSeq'].apply(is_utail)
            utailed_cnt = (subgrp['readCount'] * utailed).sum()

            nomod_cnt = (subgrp['readCount'] * (subgrp['addedSeq'] == '')).sum()
            total_cnt = subgrp['readCount'].sum()
            pos_counts.append([hairpin, endpos, total_cnt, nomod_cnt, utailed_cnt])

    endpos_tbl = pd.DataFrame(pos_counts, columns=('hairpin', 'refEndPos', 'total', 'noMod', 'uTailed'))
    endpos_tbl['uRate'] = endpos_tbl['uTailed'] / endpos_tbl['total']
    return endpos_tbl

endpos_ctl = make_modrate_tbl_by_pos(ctlreads)
endpos_kd = make_modrate_tbl_by_pos(kdreads)
```

```python
endpos_merged = pd.merge(endpos_ctl, endpos_kd, how='inner', left_on=('hairpin', 'refEndPos'),
                         right_on=('hairpin', 'refEndPos'), suffixes=('_ctl', '_kd'))
```

```python
MINIMUM_READS = 1000

endurates = []

for hairpin, grp in endpos_merged.groupby('hairpin'):
    mostfreq_pos = grp.sort('total_ctl', ascending=False).iloc[0]['refEndPos']
    if grp['total_ctl'].sum() < MINIMUM_READS:
        continue
    grp['relativePos'] = grp['refEndPos'] - mostfreq_pos
    endurates.append(grp[['relativePos', 'uRate_ctl', 'uRate_kd']])

allends = pd.concat(endurates)
endpos_summarized = allends.groupby('relativePos').agg(np.mean)
```
